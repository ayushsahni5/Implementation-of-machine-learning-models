# Training and Testing

```python
df2=pd.read_csv("automobiles3.csv")
df["price"]=df2["price"]       #re-adding price column after normalization of dataframe
df["price"].fillna(df["price"].median(),inplace=True)   #Handling NaN values in price column
df3=df.drop(["normalized-losses"],axis=1)      #i am removing whole normalized-losses column because 41 its 41 entries are null i.e. nearly 20% entrie are null. This is giving better accuracy
training_data=df3.sample(frac=0.67)
test_data= df3.drop(training_data.index)
```

```python
import math
def manhattan_distance(row1, row2):
    distance = 0.0
    for i in range(len(row1)-1):
        distance += abs(row1[i] - row2[i])
    return distance
```

```python
k=3        #k=5 was giving rmse in range 1000-3000  
for i in range(len(test_data)):
    rmse=0.0
    d_list=list()
    for j in range(len(training_data)):
        d_list.append((manhattan_distance(test_data.iloc[i],training_data.iloc[j]),training_data.iloc[j]["price"]))
    d_list.sort()
    sum=0.0
    for i in range(k):
        sum+=d_list[i][1]
    sum=sum/k    
    rmse+=(sum-test_data.iloc[i]["price"])**2
rmse=rmse/(len(test_data))    
rmse=math.sqrt(rmse)
```

```python
print(rmse)   #this is the result of model on the test_data split from the initial training data. I am assuming that
              #the initial data taken from uci is training data. I have split this into two parts. So the data
              #which I have used here for testing is nothing but validation set data. One may also assume 3 splits
              #of the initial data into training ,validation and test sets( this I have not done here) 
```

# Cross validation for hyperparameter tuning

### Now I will split the 136 rows(67% of initial rows) furthur into a new validation set

```python
#I am applying holdout cross validation method within the training data set(of 136 rows). I will keep aside 33%
#of this training data set to be used as validation set.
training_data2=training_data.sample(frac=0.67)
validation_data=training_data.drop(training_data2.index)
```

```python
import math

x_axis=list()     #for plotting
y_axis=list()     #for plotting
k=1
while(k<16):
    for i in range(len(validation_data)):   #iterating over validation set rows
        rmse=0.0
        d_list=list()                       # a list for storing distance of ith validation row with every training row
        for j in range(len(training_data2)):    #iterating over training set rows
            d_list.append((manhattan_distance(validation_data.iloc[i],training_data2.iloc[j]),training_data2.iloc[j]["price"]))
        d_list.sort()
        sum=0.0
        for i in range(k):
            sum+=d_list[i][1]
        sum=sum/k    
        rmse+=(sum-validation_data.iloc[i]["price"])**2
        rmse=rmse/(len(validation_data))    
        rmse=math.sqrt(rmse)
    x_axis.append(k)
    y_axis.append(rmse)
    k=k+2
```

```python
import matplotlib.pyplot as plot    # a plot showing value of rmse for k=1,3,5,7,9,11,13,15
fig=plot.figure()
ax=fig.add_axes([0,0,1,1])
ax.bar(x_axis,y_axis)
plot.show()
```
