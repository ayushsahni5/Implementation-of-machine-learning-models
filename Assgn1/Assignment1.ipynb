# Training and Testing

```python
df2=pd.read_csv("automobiles3.csv")
df["price"]=df2["price"]       #re-adding price column after normalization of dataframe
df["price"].fillna(df["price"].median(),inplace=True)   #Handling NaN values in price column
df3=df.drop(["normalized-losses"],axis=1)
training_data=df.sample(frac=0.67)
test_data= df.drop(training_data.index)
```

```python
import math
def euclidean_distance(row1, row2):
    distance = 0.0
    for i in range(len(row1)-1):
        distance += abs(row1[i] - row2[i])
    return distance
```

```python
k=3        #k=5 was giving rmse in range 1000-3000  
for i in range(len(test_data)):
    rmse=0.0
    d_list=list()
    for j in range(len(training_data)):
        d_list.append((euclidean_distance(test_data.iloc[i],training_data.iloc[j]),training_data.iloc[j]["price"]))
    d_list.sort()
    sum=0.0
    for i in range(k):
        sum+=d_list[i][1]
    sum=sum/k    
    rmse+=(sum-test_data.iloc[i]["price"])**2
rmse=rmse/(len(test_data))    
rmse=math.sqrt(rmse)
```
